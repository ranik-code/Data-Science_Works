```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('melbourne_housing.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
cols_to_zero = ['Distance','Bedroom2', 'Bathroom', 'Car', 'Propertycount']
df[cols_to_zero] = df[cols_to_zero].fillna(0)
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df = df.drop(columns=['Address', 'Date', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude'])
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.columns
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
X = df.drop(columns=['Price'])
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, Y_train)
```

```python
print('Train Score : \t', reg.score(X_train, Y_train))
print('Test Score : \t', reg.score(X_test, Y_test))
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=100, max_iter=100)
lasso_reg.fit(X_train, Y_train)
```

```python
print('Train Score : \t', lasso_reg.score(X_train, Y_train))
print('Test Score : \t', lasso_reg.score(X_test, Y_test))
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=100, max_iter=100)
ridge_reg.fit(X_train, Y_train)
```

```python
print('Train Score : \t', ridge_reg.score(X_train, Y_train))
print('Test Score : \t', ridge_reg.score(X_test, Y_test))
```
