## Q4. Calculate the mean/median, variance, standard deviation of the above data using statistics module, using your own function, using numpy.

```python
def Mean(df) :
    print("Mean :")
    for col in df.columns:
        x = df[col].tolist()
        print(col,':', stat.mean(x), sep='\t')
        
def Median(df) :
    print("Median :")
    for col in df.columns:
        x = df[col].tolist()
        print(col,':', stat.median(x), sep='\t')
        
def Variance(df) :
    print("Variance :")
    for col in df.columns:
        x = df[col].tolist()
        print(col,':', stat.variance(x), sep='\t')
        
def StandardDeviation(df) :
    print("Standard Deviation :")
    for col in df.columns:
        x = df[col].tolist()
        print(col,':', stat.stdev(x), sep='\t')
```

## Q5. Calculate Euclidean Distance.

```python
def Euclidean_Distance(x, y):
 return sum([ (x_i - y_i) ** 2 for x_i, y_i in zip(x, y) ]) ** 0.5
```

```python
cols = A.iloc[:,1:].columns #A = Input Labels of "Iris" Dataset

for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        a = cols[i]
        b = cols[j]
        print(a, b, Euclidean_Distance(df1[a], df1[b]), sep='\t')
```

```python
cols2 = C.iloc[:,1:-1].columns #C = Input Labels of "Breast Cancer" Dataset
for i in range(len(cols2)):
     for j in range(i+1, len(cols2)):
        a = cols2[i]
        b = cols2[j]
        print(a, b, Euclidean_Distance(df2[a], df2[b]), sep='\t')
```
